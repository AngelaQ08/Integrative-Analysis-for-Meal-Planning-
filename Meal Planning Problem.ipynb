```python
nutri = f[["title", "calories", "sodium", "fat", "protein"]]
nutri
```

```python
B = f[["title", "Price"]]
B
```

```python
##reduce people version
g = {}
for i in range(len(pr)):
    g[i] = pr["personal_rating"][i]
g = pd.DataFrame(dict([(k,pd.Series(v)) for k,v in g.items() ]))
g.columns = pr.title
g = pd.melt(g)
g = g.dropna()
g = g.join(pd.DataFrame(g.pop('value').values.tolist(), index=g.index, columns = ["person_info", "person rating"]))
new = g['person_info'].str.split("from",n = 1, expand = True)
g["person"] = new[0]
g["Location"] = new[1]
g['count'] = g['person'].map(g['person'].value_counts())
g = g.drop(columns = ["person_info", "Location"])
g = g.drop(g.loc[g["count"] < 4].index).reset_index()
#g = g.drop(columns = ["count"])
m = g.pivot_table(index = "person", columns = "title", values = "person rating")
m.index = pd.Series(range(1,len(m)+1))
#m.to_csv("p1ratingdata.csv")
m
```

```python
final = pd.merge(g[["title"]], nutri, on = 'title')
final = pd.merge(final, B, on = 'title')
final = final.drop_duplicates()
final = final.sort_values("title")
final = final.reset_index(drop = True)
final
```
